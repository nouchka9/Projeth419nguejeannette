```python
import os
import shutil
import random
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, TensorBoard
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import seaborn as sns
import numpy as np
from tensorflow.keras.preprocessing import image
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="keras")

# Masquer les messages INFO et WARNING de TensorFlow
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"

# Activation du "memory growth" sur le GPU (si disponible)
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# =============================================================
# 1. ParamÃ¨tres et rÃ©pertoires de base
# =============================================================
# Dans Kaggle, le dataset est dÃ©jÃ  attachÃ© sous /kaggle/input
BASE_DIR = "/kaggle/input/chest-xray-pneumonia/chest_xray"
WORKING_DIR = "/kaggle/working/chest_xray"
NEW_DIR = "/kaggle/working/repartition_chest_xray"  # Nouveau rÃ©pertoire de rÃ©partition (80/10/10)

# =============================================================
# 2. CrÃ©ation du rÃ©pertoire de travail et copie du dataset original
# =============================================================
def create_working_directory(BASE_DIR, WORKING_DIR):
    """
    Copie intÃ©gralement le contenu de BASE_DIR dans WORKING_DIR pour pouvoir le modifier.
    """
    if not os.path.exists(WORKING_DIR):
        shutil.copytree(BASE_DIR, WORKING_DIR)
    print("âœ… Copie du dataset dans WORKING_DIR terminÃ©e.")

# =============================================================
# 3. Fonctions utilitaires pour le comptage et la vÃ©rification de la structure
# =============================================================
def count_images(directory):
    """
    Compte le nombre d'images (extensions .jpg, .jpeg, .png, .bmp) dans le rÃ©pertoire 'directory' de maniÃ¨re rÃ©cursive.
    """
    valid_ext = {".jpg", ".jpeg", ".png", ".bmp"}
    count = 0
    for root, dirs, files in os.walk(directory):
        for f in files:
            if os.path.splitext(f)[1].lower() in valid_ext:
                count += 1
    return count

def check_structure(base_dir):
    """
    Affiche la liste des partitions (train, val, test) prÃ©sentes dans base_dir et le nombre d'images dans chacune.
    """
    partitions = ["train", "val", "test"]
    current_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    print("=== Structure de WORKING_DIR ===")
    print("Dossiers prÃ©sents :", current_dirs)
    for part in partitions:
        part_path = os.path.join(base_dir, part)
        if os.path.exists(part_path):
            n = count_images(part_path)
            print(f" - {part.upper()} : {n} images")
        else:
            print(f" - {part.upper()} : dossier non trouvÃ©")
    print("âœ… VÃ©rification de la structure terminÃ©e.\n")

# =============================================================
# 4. Repartitionnement global en 80% TRAIN, 10% VAL, 10% TEST
# =============================================================
def repartition_dataset_by_ratio(source_dir, target_dir, ratio_train=0.8, ratio_val=0.1, ratio_test=0.1):
    """
    Pour chaque classe ("NORMAL" et "PNEUMONIA"), regroupe toutes les images prÃ©sentes dans les partitions
    existantes (train, val, test) de source_dir, puis les rÃ©partit dans target_dir selon les proportions :
      - TRAIN : ratio_train
      - VAL   : ratio_val
      - TEST  : ratio_test
      
    Le nombre total d'images est conservÃ©.
    Affiche le compte d'images pour chaque classe avant et aprÃ¨s repartition.
    """
    classes = ["NORMAL", "PNEUMONIA"]
    
    # CrÃ©ation des sous-dossiers cibles
    for part in ["train", "val", "test"]:
        for cls in classes:
            os.makedirs(os.path.join(target_dir, part, cls), exist_ok=True)
    
    for cls in classes:
        all_images = []
        # Pour chaque partition existante (train, val, test) dans source_dir, rÃ©cupÃ©rer les images de la classe cls
        for part in ["train", "val", "test"]:
            part_dir = os.path.join(source_dir, part, cls)
            if os.path.exists(part_dir):
                for root, dirs, files in os.walk(part_dir):
                    for file in files:
                        if os.path.splitext(file)[1].lower() in {".jpg", ".jpeg", ".png", ".bmp"}:
                            all_images.append(os.path.join(root, file))
        total = len(all_images)
        print(f"[{cls}] Images totales dans le dataset d'origine : {total}")
        
        # Calcul du nombre dÃ©sirÃ© dans chaque partition
        n_train = int(total * ratio_train)
        n_val   = int(total * ratio_val)
        n_test  = total - n_train - n_val  # Pour garantir la somme totale
        
        print(f"[{cls}] Nouvelle rÃ©partition souhaitÃ©e -> TRAIN: {n_train}, VAL: {n_val}, TEST: {n_test}")
        
        # Shuffle et rÃ©partition
        random.shuffle(all_images)
        train_list = all_images[:n_train]
        val_list   = all_images[n_train:n_train + n_val]
        test_list  = all_images[n_train + n_val:]
        
        # Copier les images dans les dossiers cibles
        for img_path in train_list:
            shutil.copy(img_path, os.path.join(target_dir, "train", cls, os.path.basename(img_path)))
        for img_path in val_list:
            shutil.copy(img_path, os.path.join(target_dir, "val", cls, os.path.basename(img_path)))
        for img_path in test_list:
            shutil.copy(img_path, os.path.join(target_dir, "test", cls, os.path.basename(img_path)))
    
    # Afficher le compte d'images aprÃ¨s repartition
    print("\n=== RÃ©partition finale dans target_dir ===")
    for part in ["train", "val", "test"]:
        total_part = 0
        for cls in classes:
            cnt = len(os.listdir(os.path.join(target_dir, part, cls)))
            total_part += cnt
            print(f" - {part.upper()} [{cls}] : {cnt} images")
        print(f"=> Total {part.upper()} : {total_part} images\n")

# =============================================================
# 5. CrÃ©ation des GÃ©nÃ©rateurs d'Images
# =============================================================
def create_generators(TRAIN_DIR, VAL_DIR, TEST_DIR, target_size=(150, 150), batch_size=32):
    """
    CrÃ©e et retourne les gÃ©nÃ©rateurs d'images pour TRAIN, VAL et TEST Ã  partir de target_dir.
    Le gÃ©nÃ©rateur de test est crÃ©Ã© avec shuffle=False pour conserver l'ordre.
    """
    train_datagen = ImageDataGenerator(
        rescale=1.0/255,
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        zoom_range=0.2,
        shear_range=0.1,
        horizontal_flip=True
    )
    val_test_datagen = ImageDataGenerator(rescale=1.0/255)
    
    train_generator = train_datagen.flow_from_directory(
        TRAIN_DIR,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='binary'
    )
    
    val_generator = val_test_datagen.flow_from_directory(
        VAL_DIR,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='binary'
    )
    
    test_generator = val_test_datagen.flow_from_directory(
        TEST_DIR,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='binary',
        shuffle=False
    )
    return train_generator, val_generator, test_generator

# =============================================================
# 6. Construction et EntraÃ®nement du ModÃ¨le CNN
# =============================================================
def build_model(input_shape=(150, 150, 3)):
    """
    Construit et compile un modÃ¨le CNN simple.
    """
    model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3,3), activation='relu'),
        MaxPooling2D(pool_size=(2,2)),
        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D(pool_size=(2,2)),
        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D(pool_size=(2,2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

def plot_training_history(history):
    """
    Affiche les courbes d'apprentissage (perte et prÃ©cision).
    """
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(history.history['loss'], label='Perte EntraÃ®nement', color='blue')
    plt.plot(history.history['val_loss'], label='Perte Validation', color='orange')
    plt.title("Courbe de Perte")
    plt.xlabel("Ã‰poques")
    plt.ylabel("Perte")
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(history.history['accuracy'], label='PrÃ©cision EntraÃ®nement', color='blue')
    plt.plot(history.history['val_accuracy'], label='PrÃ©cision Validation', color='orange')
    plt.title("Courbe de PrÃ©cision")
    plt.xlabel("Ã‰poques")
    plt.ylabel("PrÃ©cision")
    plt.legend()
    plt.show()

def plot_combined_history(history):
    """
    Affiche conjointement les courbes de perte et de prÃ©cision sur deux axes.
    """
    fig, ax1 = plt.subplots(figsize=(8,6))
    ax1.set_xlabel("Ã‰poques")
    ax1.set_ylabel("Perte", color='blue')
    ax1.plot(history.history['loss'], label='Perte EntraÃ®nement', color='blue', linestyle='dashed')
    ax1.plot(history.history['val_loss'], label='Perte Validation', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    
    ax2 = ax1.twinx()
    ax2.set_ylabel("PrÃ©cision", color='orange')
    ax2.plot(history.history['accuracy'], label='PrÃ©cision EntraÃ®nement', color='orange', linestyle='dashed')
    ax2.plot(history.history['val_accuracy'], label='PrÃ©cision Validation', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    
    fig.suptitle("Courbes combinÃ©es de Perte et PrÃ©cision")
    fig.legend(loc="upper right")
    plt.show()

# =============================================================
# 7. Ã‰valuation du ModÃ¨le (Matrice de Confusion et Rapport de Classification)
# =============================================================
def evaluate_model(model, test_generator):
    """
    Ã‰value le modÃ¨le sur le set de test, affiche la matrice de confusion et le rapport de classification.
    """
    test_loss, test_acc = model.evaluate(test_generator)
    print(f"ğŸ¯ Performance sur le set de test :")
    print(f"   â¡ï¸ Test Accuracy : {test_acc:.4f}")
    print(f"   â¡ï¸ Test Loss     : {test_loss:.4f}")
    
    y_pred = (model.predict(test_generator) > 0.5).astype("int32")
    y_true = test_generator.classes
    
    # Affichage de la matrice de confusion
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=['NORMAL', 'PNEUMONIA'], yticklabels=['NORMAL', 'PNEUMONIA'])
    plt.title("Matrice de Confusion")
    plt.xlabel("PrÃ©dictions")
    plt.ylabel("VÃ©ritÃ©s")
    plt.show()
    
    # Affichage du rapport de classification
    report = classification_report(y_true, y_pred, target_names=['NORMAL', 'PNEUMONIA'])
    print("\nğŸ“Š Rapport de Classification :\n", report)
    
    return y_true, y_pred

def plot_roc_curve(model, test_generator):
    """
    Calcule et affiche la courbe ROC avec l'AUC.
    """
    y_probs = model.predict(test_generator).ravel()
    fpr, tpr, thresholds = roc_curve(test_generator.classes, y_probs)
    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {roc_auc:.2f})')
    plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Taux de Faux Positifs')
    plt.ylabel('Taux de Vrais Positifs')
    plt.title('Courbe ROC')
    plt.legend(loc="lower right")
    plt.show()

# =============================================================
# 8. PrÃ©diction sur quelques Images et Affichage
# =============================================================
def predict_and_plot_images(model, test_generator, num_images=10, target_size=(150,150)):
    """
    SÃ©lectionne alÃ©atoirement 'num_images' dans le set de test et affiche pour chacune :
      - Le label rÃ©el,
      - La prÃ©diction du modÃ¨le,
      - Le niveau de confiance.
    """
    idx2class = {v: k for k, v in test_generator.class_indices.items()}
    test_image_paths = test_generator.filepaths
    selected_images = random.sample(test_image_paths, num_images)
    
    plt.figure(figsize=(15,6))
    for i, img_path in enumerate(selected_images):
        img_idx = test_generator.filepaths.index(img_path)
        true_label = idx2class.get(test_generator.classes[img_idx], "Inconnu")
        
        img = image.load_img(img_path, target_size=target_size)
        img_array = image.img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)
        
        pred = model.predict(img_array)[0][0]
        confidence = pred if pred > 0.5 else 1 - pred
        predicted_class = "PNEUMONIA" if pred > 0.5 else "NORMAL"
        title_text = f"RÃ©el: {true_label}\nPrÃ©vu: {predicted_class}\nConfiance: {confidence:.2%}"
        
        plt.subplot(2, num_images // 2, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(title_text, fontsize=10)
        
    plt.suptitle("PrÃ©dictions sur des images alÃ©atoires du set de test", fontsize=14)
    plt.show()

# =============================================================
# Fonction Principale
# =============================================================
def main():
    # 1. Copie du dataset attachÃ© de Kaggle vers WORKING_DIR
    # BASE_DIR est le chemin fourni par Kaggle.
    BASE_DIR = "/kaggle/input/chest-xray-pneumonia/chest_xray"
    if not os.path.exists(BASE_DIR):
        raise ValueError("Le dossier de dataset BASE_DIR est introuvable !")
    
    if not os.path.exists(WORKING_DIR):
        create_working_directory(BASE_DIR, WORKING_DIR)
    else:
        print("âš ï¸ WORKING_DIR existe dÃ©jÃ , utilisation du contenu existant.")
    
    # Afficher la rÃ©partition d'origine (avant repartition)
    print("=== RÃ©partition d'origine (avant repartition) ===")
    check_structure(WORKING_DIR)
    
    # 2. DÃ©finition des chemins pour TRAIN, VAL et TEST dans WORKING_DIR
    TRAIN_DIR = os.path.join(WORKING_DIR, "train")
    TEST_DIR = os.path.join(WORKING_DIR, "test")
    VAL_DIR = os.path.join(WORKING_DIR, "val")
    if not os.path.exists(VAL_DIR):
        os.makedirs(VAL_DIR)
    
    # 3. Repartir globalement le dataset en 80% TRAIN, 10% VAL et 10% TEST
    # Cette Ã©tape regroupe toutes les images issues de TRAIN, VAL et TEST
    # pour chaque classe ("NORMAL" et "PNEUMONIA") puis les rÃ©partit selon les proportions souhaitÃ©es.
    if os.path.exists(NEW_DIR):
        shutil.rmtree(NEW_DIR)
    os.makedirs(NEW_DIR, exist_ok=True)
    repartition_dataset_by_ratio(WORKING_DIR, NEW_DIR, ratio_train=0.8, ratio_val=0.1, ratio_test=0.1)
    
    # Afficher la nouvelle rÃ©partition (aprÃ¨s repartition)
    print("=== RÃ©partition finale (aprÃ¨s repartition) ===")
    check_structure(NEW_DIR)
    
    # 4. CrÃ©ation des gÃ©nÃ©rateurs d'images Ã  partir du dossier repartitionnÃ©
    TRAIN_NEW = os.path.join(NEW_DIR, "train")
    VAL_NEW = os.path.join(NEW_DIR, "val")
    TEST_NEW = os.path.join(NEW_DIR, "test")
    train_gen, val_gen, test_gen = create_generators(TRAIN_NEW, VAL_NEW, TEST_NEW)
    print(f"ğŸ“Š Images d'entraÃ®nement : {train_gen.samples}")
    print(f"ğŸ“Š Images de validation  : {val_gen.samples}")
    print(f"ğŸ“Š Images de test        : {test_gen.samples}")
    total_images = train_gen.samples + val_gen.samples + test_gen.samples
    print(f"ğŸ“Š Total images rÃ©parties : {total_images}")
    
    # 5. Construction et entraÃ®nement du modÃ¨le CNN
    model = build_model(input_shape=(150, 150, 3))
    model.summary()
    
    # DÃ©finition des callbacks
    lr_reduction = ReduceLROnPlateau(monitor='val_loss', patience=3, verbose=1, factor=0.5, min_lr=1e-7)
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
    tensorboard_cb = TensorBoard(log_dir="./logs")
    callbacks_list = [lr_reduction, early_stop, tensorboard_cb]
    
    history = model.fit(
        train_gen,
        epochs=30,
        validation_data=val_gen,
        callbacks=callbacks_list
    )
    print("âœ… EntraÃ®nement terminÃ© avec succÃ¨s.")
    
    # 6. Ã‰valuation et visualisation
    evaluate_model(model, test_gen)
    plot_training_history(history)
    plot_combined_history(history)
    plot_roc_curve(model, test_gen)
    predict_and_plot_images(model, test_gen, num_images=10)

if __name__ == "__main__":
    main()
```
